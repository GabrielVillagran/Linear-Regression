#Diabetes Example
Loading the data and importing the neccessary libraries 

```python
# Code source: Jaques Grobler
# License: BSD 3 clause
#Import the neccessary libraries 
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

# Load the diabetes dataset
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)

# Use only one feature
diabetes_X = diabetes_X[:, np.newaxis, 2]
size_x = len(diabetes_X)
size_y = len(diabetes_y)
```

Total size of the data (X and Y axis)

```python
print("The total data size of X: ", size_x)
print("The total data size of Y: ", size_y)
```

Calculating the percent for each axis 

```python
#Data on X axis 
#train
desc = size_x * (80/100)
train = round(size_x-desc)
print("Training(X) = ",train)
#test
desc = size_x * (20/100)
test = round(size_x-desc)
print("Test(X)= ", test)
print("Training + Test = ", train + test)
print("===================")

#Data on Y axis 
#train
desc = size_y * (80/100)
train = round(size_y-desc)
print("Training(Y) = ",train)
#test
desc = size_y * (20/100)
test = round(size_y-desc)
print("Test(Y)= ", test)
print("Training + Test = ", train + test)
```

#Splitting the data into **Training = 80%** and **Test = 20%**
First I got the values required for each axis and then I started to search the required value for each axis and its respective set (train or test).
I found that we can put values on bot side and that was much easier to find the values.

```python
# Split the data into training/testing sets
diabetes_X_train = diabetes_X[-88:]
diabetes_X_test = diabetes_X[:-88]
print("Training(X): ",len(diabetes_X_train))
print("Test(X): ",len(diabetes_X_test))

# Split the targets into training/testing sets
diabetes_y_train = diabetes_y[-88:]
diabetes_y_test = diabetes_y[:-88]
print("Training(y): ",len(diabetes_y_train))
print("Test(y): ",len(diabetes_y_test))
```

```python
# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(diabetes_X_train, diabetes_y_train)

# Make predictions using the testing set
diabetes_y_pred = regr.predict(diabetes_X_test)

# The coefficients
print("Coefficients: \n", regr.coef_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(diabetes_y_test, diabetes_y_pred))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(diabetes_y_test, diabetes_y_pred))

# Plot outputs
plt.scatter(diabetes_X_test, diabetes_y_test, color="red")
plt.plot(diabetes_X_test, diabetes_y_pred, color="blue", linewidth=2)

plt.xticks(())
plt.yticks(())

plt.show()
```

#Splitting the data into **Training = 80%** and **Test = 20%**
